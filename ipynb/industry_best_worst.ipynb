## 1. Cleaning our csv data

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
# import all relevant csvs/xlxs into dataframes for analysis

#collect specific columns for each df
epa_viols_df = pd.read_csv("../datasets/epa_violations.csv", 
                           usecols=[0,1,2,3,4,5,6,25,26,28,29], index_col = 0)

scope_emis_df = pd.read_csv("../datasets/2013_-_Global_500_Emissions_and_Response_Status.csv", 
                            usecols=[0,2,4,11,12], index_col = 0)
just_co_df = pd.read_excel('../datasets/JUST Capital 2018 Company Scores.xlsx')
```

```python
# clean our data and sort by stock ticker
epa_viol_st = epa_viols_df.dropna(subset=['Parent Company Stock Ticker'])
epa_viol_st = epa_viol_st.sort_values("Parent Company Stock Ticker")
epa_viol_st['Penalty Date'] = epa_viol_st['Penalty Date'].apply(
    lambda date: str(date)[4:6]+"/"+str(date)[6:])

#change penalties to int
epa_viol_st['Penalty Amount'] = epa_viol_st['Penalty Amount'].apply(
    lambda amt: int(re.sub('[$,]','',amt)))
epa_viol_st['Subtraction From Penalty'] = epa_viol_st['Subtraction From Penalty'].apply(
    lambda amt: int(re.sub('[$,]','',amt)))
epa_viol_st['Penalty Amount Adjusted For Eliminating Multiple Counting'] = epa_viol_st['Penalty Amount Adjusted For Eliminating Multiple Counting'].apply(
    lambda amt: int(re.sub('[$,]','',amt)))
epa_viol_st.columns = ['Parent Company','Penalty ($)','Sub From Penalty ($)', 'Penalty Amount Adjusted For Eliminating Multiple Counting ($)','Year','Date','HQ Country','HQ State','Ticker','Parent Industry']
# print(len(epa_viol_st['Parent Company Stock Ticker'].unique()))
# convert list of all tickers into a set
epa_cos_set = set(epa_viol_st['Ticker'])
# len(epa_cos_set)
epa_viol_st[:50]
# print(ope_so)
```

```python
# do the same for scope_emis_df
# we need to remove the country code for comparison
scope_emis_df.columns = ['Country', 'Ticker', 'Scope 1 (metric tons CO2e)', 'Scope 2 (metric tons CO2e)']
scope_emis_df['Ticker'] = scope_emis_df['Ticker'].apply(
    lambda sym: sym.split()[0])
scope_emis_df = scope_emis_df.dropna(subset=['Scope 1 (metric tons CO2e)','Scope 2 (metric tons CO2e)'])
scope_cos_set = set(scope_emis_df['Ticker'])
print(len(scope_cos_set))
scope_emis_df[:50]
```

```python
# lastly for just_co_df

#find only relevant columns which are Ticker, Name, Industry, and ENV scores
jcols = ['TICKER', 'NAME', 'JUST_IND']
for c in just_co_df.columns:
    if c.startswith('ENV'):
        jcols.append(c)
        
# #apply new columns to table
just_co_new_df = just_co_df[jcols].copy()

# #rename columns 
just_co_new_df.columns = ['Ticker', 
                  'Name', 
                  'Industry', 
                  'Total Score', 
                  'Efficiency', #Uses natural resources efficently and reduces product-related waste.
                  'Management', #Has policies and practices to protect the environment
                  'Pollution', #Minimizes pollution and cleans up any environmental damage it causes.
                  'E.Clean', #Maximizes Use of Renewable Energy
                  'E.Waste', #Reduces Solid Waste to Landfill
                  'P.GHG', #Greenhouse Gas Emissions
                  'P.Chem', #Chemical Water Discharge
                  'P.RSEI', #Toxic Chemical Emissions (Risk-Screening Environmental Indicators )
                  'M.Mgmt', #Environmental Management Systems and Disclosure
                  'M.Controversy', #Controversies in Environmental Responsibility as Reported in the Media
                  'M.Fines' #Environmental Fines
                 ]
just_co_set = set(just_co_new_df['Ticker'])
print(len(just_co_set))
just_co_new_df[:50]
```

## 2. Scrape for Market Cap

```python
intersection_all_dfs = just_co_set & scope_cos_set & epa_cos_set
intersn_epa_just = epa_cos_set & just_co_set
intersn_epa_scope = scope_cos_set & epa_cos_set
intersn_just_scope = just_co_set & scope_cos_set
print("{} similar companies between all 3 datasets".format(len(intersection_all_dfs)))
print("{} similar companies between EPA Violations & Scope 1&2 Cos datasets".format(len(intersn_epa_scope)))
print("{} similar companies between EPA Violations & Just Cos datasets".format(len(intersn_epa_just)))
print("{} similar companies between Scope 1&2 Cos & Just Cos datasets".format(len(intersn_just_scope)))
```

```python
# with all 3 of these datasets combined, theres 110 stock companies that are the same!
# we will now scrape yahoo finance to get the estimated valuation of the corporation/company
# In this case, we will use the Market Cap, which helps estimate the value of a company
from bs4 import BeautifulSoup as bs
import requests
import time
url_a = 'https://finance.yahoo.com/quote/'
url_b = '/key-statistics?p='
#use AA to demo
html = requests.get(url_a+'AA'+url_b+'AA')
bs_obj = bs(html.text)
table = bs_obj.find('table',class_ = 'table-qsp-stats Mt(10px)')
trs = table.tbody.findChildren(recursive = False)

for tr in trs:
    if(tr.td.span.text == 'Market Cap (intraday)'):
        print(tr.find('td', class_ = 'Fz(s) Fw(500) Ta(end)').text)
        break
```

```python
# create a dictionary with market values and ticker
company_value = {
    "Ticker":[],
    "Company Valuation (Millions)":[]
}

# We will now automate this scrape so we can get all 84 companies 
# We will use a 1.5 sec timer between requests to not overload the server
# and then add into a dictionary to create a dataframe
for ticker in intersection_all_dfs:
    time.sleep(1.5)
    company_value['Ticker'].append(ticker)
    url = 'https://finance.yahoo.com/quote/{0}/key-statistics?p={0}'.format(ticker)
    html = requests.get(url)
    bs_obj = bs(html.text)
    table = bs_obj.find('table',class_ = 'table-qsp-stats Mt(10px)')
    trs = table.tbody.findChildren(recursive = False)
    for tr in trs:
        try:
            if(tr.td.span.text == 'Market Cap (intraday)'):
                cw = tr.find('td', class_ = 'Fz(s) Fw(500) Ta(end)').text
                # Notice we have B and M values in our cw, we need to convert to numbers
                if cw == 'N/A':
                    cw = 0
                else:
                    if cw[-1] == "M":
                        cw = float(cw[:-1])
                    elif cw[-1] == 'B':
                        cw = float(cw[:-1])*1000
                company_value['Company Valuation (Millions)'].append(cw)
                break
        except AttributeError:
            company_value['Company Valuation (Millions)'].append(np.nan)
            print(ticker)
            break

# Convert to a df
company_valuation_df = pd.DataFrame(company_value)
```

```python
# save this as a csv to reference later if needed
company_valuation_df.to_csv("../datasets/market_cap_with_ticker.csv", index = False)
```

```python
company_valuation_df.sort_values('Company Valuation (Millions)')
```

```python
# lets clean the epa violations so that we only have ticker, parent co, penalties, HQ Country, industry
cleaned_epa = epa_viol_st.groupby(['Ticker', 'Parent Company','HQ Country','Parent Industry']).agg(
    {'Penalty ($)':'sum'}).reset_index()
cleaned_epa[:50]
```

```python
# merge of epa violations and stocks merge
epa_stk_merge1 = pd.concat([cleaned_epa.set_index('Ticker'),company_valuation_df.set_index('Ticker')],
                            axis=1, join='inner').reset_index()

# merge of merge1 and scope
scope_merge1_2 = pd.merge(epa_stk_merge1, scope_emis_df[['Ticker','Scope 1 (metric tons CO2e)','Scope 2 (metric tons CO2e)']],
                         on='Ticker')

# merge of just cap and merge2
complete_merge_df = pd.merge(scope_merge1_2, just_co_new_df[['Ticker','Total Score']],
                            on='Ticker')

# Add a new column to understand relative size based off company valuation
largest_co = complete_merge_df['Company Valuation (Millions)'].max()
complete_merge_df['Relative Size'] = complete_merge_df['Company Valuation (Millions)'].apply(
    lambda val: (val/largest_co)*100)
```

## 3. Plotly Graphs

```python
import plotly.plotly as py
import plotly.graph_objs as go

# initialize var we will use 
# hover text
bubble_text = []

#for gradient of bubble
hi_scope1 = complete_merge_df['Scope 1 (metric tons CO2e)'].max()
low_scope1 = complete_merge_df['Scope 1 (metric tons CO2e)'].max()
cmpl_no_out = complete_merge_df[~complete_merge_df['Ticker'].isin(['APC','XOM', 'CVX'])]
```

```python
# lets create a scatter plot of all our data using plotly



# this is so we can add additional information for when we hover over a bubble
bubble_text = []
for _, row in complete_merge_df.iterrows():
    bubble_text.append(('Ticker: {tick}<br>'
                       + 'Company: {co}<br>'
                       + 'Valuation: ${val}M<br>'
                       + 'Country: {hq}<br>'
                       + 'Industry: {ind}<br>'
                       + 'Scope 1: {s1}<br>'
                       + 'Scope 2: {s2}').format(tick=row['Ticker'],
                                                co=row['Parent Company'],
                                                val=row['Company Valuation (Millions)'],
                                                hq=row['HQ Country'],
                                                ind=row['Parent Industry'],
                                                s1=row['Scope 1 (metric tons CO2e)'],
                                                s2=row['Scope 2 (metric tons CO2e)']))
# add this into the dataframe as another column
complete_merge_df['text_display'] = bubble_text

# Trace 0 represents our first scatter. we define our x and y axis here as well
# as other info such as color, marker type, hover text
trace0 = go.Scatter(
    x = complete_merge_df['Total Score'],
    y = complete_merge_df['Penalty ($)'],
    mode = 'markers',
    text = complete_merge_df['text_display'],
    marker = dict(
        size = complete_merge_df['Relative Size'],
        cmin = low_scope1,
        cmax = hi_scope1,
        color = complete_merge_df['Scope 1 (metric tons CO2e)'],
        colorbar = dict(
            title = 'Scope 1 Emissions'
        ),
        colorscale='Viridis'
    )
)

# Add this into our data that our figure will read to populate the graph
data = [trace0]

# This layout will include x and y axis labels and our title
layout = go.Layout(
    title = 'Just Score vs EPA Penalty',
    xaxis=dict(
        title = 'Just Score'
    ),
    yaxis=dict(
        title = 'EPA Penalty (Dollars)'
    )
)

# plots
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Just Score vs EPA Penalty')
```

```python
# The top diagram is great, but we see 3 outliers, APC, XOM, and CVX.
# Let's remove these and replot

# this is so we can add additional information for when we hover over a bubble
bubble_text = []
for _, row in cmpl_no_out.iterrows():
    bubble_text.append(('Ticker: {tick}<br>'
                       + 'Company: {co}<br>'
                       + 'Valuation: ${val}M<br>'
                       + 'Country: {hq}<br>'
                       + 'Industry: {ind}<br>'
                       + 'Scope 1: {s1}<br>'
                       + 'Scope 2: {s2}').format(tick=row['Ticker'],
                                                co=row['Parent Company'],
                                                val=row['Company Valuation (Millions)'],
                                                hq=row['HQ Country'],
                                                ind=row['Parent Industry'],
                                                s1=row['Scope 1 (metric tons CO2e)'],
                                                s2=row['Scope 2 (metric tons CO2e)']))
cmpl_no_out['text_display'] = bubble_text

trace0 = go.Scatter(
    x = cmpl_no_out['Total Score'],
    y = cmpl_no_out['Penalty ($)'],
    mode = 'markers',
    text = cmpl_no_out['text_display'],
    marker = dict(
        size = cmpl_no_out['Relative Size'],
        cmin = low_scope1,
        cmax = hi_scope1,
        color = cmpl_no_out['Scope 1 (metric tons CO2e)'],
        colorbar = dict(
            title = 'Scope 1 Emissions'
        ),
        colorscale='Viridis'
    )
)

data = [trace0]

layout = go.Layout(
    title = 'Just Score vs EPA Penalty',
    xaxis=dict(
        title = 'Just Score'
    ),
    yaxis=dict(
        title = 'EPA Penalty (Dollars)'
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Just Score vs EPA Penalty no outliers')
```

```python
# take a look at different industries. So lets create this as a function so we can reuse code
def plotly_fig(df, industry, ttl):
    trace0 = go.Scatter(
        x = df['Total Score'][df['Parent Industry'] == industry],
        y = df['Penalty ($)'][df['Parent Industry'] == industry],
        mode = 'markers',
        text = df['text_display'][df['Parent Industry'] == industry],
        marker = dict(
            size = df['Relative Size'][df['Parent Industry'] == industry],
            cmin = low_scope1,
            cmax = hi_scope1,
            color = df['Scope 1 (metric tons CO2e)'][df['Parent Industry'] == industry],
            colorbar = dict(
                title = 'Scope 1 Emissions'
            ),
            colorscale='Viridis'
        )
    )

    data = [trace0]

    layout = go.Layout(
        title = 'Just Score vs EPA Penalty by Industry: ' + ttl,
        xaxis=dict(
            title = 'Just Score'
        ),
        yaxis=dict(
            title = 'EPA Penalty (Dollars)'
        )
    )
    return go.Figure(data=data, layout=layout)
```

```python
# Take a look at pharma, oil&gas, retailing, financial services
fig_pharma = plotly_fig(complete_merge_df, "pharmaceuticals", "Pharmaceuticals")
py.iplot(fig_pharma, filename='Just Score vs EPA Penalty Pharmaceuticals')
```

```python
fig_oilgas = plotly_fig(complete_merge_df, "oil and gas", 'Oil & Gas')
py.iplot(fig_oilgas, filename='Just Score vs EPA Penalty Oil & Gas')
```
